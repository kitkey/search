```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("./search_time.csv")
algos= ['linear', 'bst', 'rbt', 'hash', 'multimap']

fig, axes = plt.subplots(nrows=5, ncols=1, figsize=(8, 12), sharex=True)
for ax, alg in zip(axes, algos):
    sub = data[data['algorithm'] == alg]
    ax.plot(sub['size'], sub['time'], marker='o')
    ax.set_title(f"{alg} search")
    ax.set_ylabel("Время, с")
    ax.grid(True)

axes[-1].set_xlabel("Размер")
plt.tight_layout()
plt.show()
```

```python
hash_data = data[data['algorithm'] == 'hash']

plt.figure(figsize=(8, 5))
plt.plot(hash_data['size'], hash_data['collision'], marker='o')
plt.xlabel("Размер массива")
plt.ylabel("Количество коллизий")
plt.title("Коллизии хэш-таблицы")
plt.grid(True)
plt.tight_layout()
plt.show()
```
